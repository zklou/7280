```python
# Write a function that returns the 3 10-node toy networks from the assignment pdf

def create_toy_graphs() -> Tuple[nx.Graph, nx.Graph, nx.Graph]:
    
    """
    
    Returns:
    cycle: a networkx graph object meeting the requirements of a cycle
    clique: a networkx graph object meeting the requirements of a clique
    star: a networkx graph object meeting the requirements of a star network
    
    """
    
    # Cycle network: A 10-node cycle network
    cycle = nx.cycle_graph(10)
    
    # Clique network: A complete graph with 10 nodes
    clique = nx.complete_graph(10)
    
    # Star network: A star network with 10 nodes (1 central node connected to 9 peripheral nodes)
    star = nx.star_graph(9)  # The center node is node 0, with 9 peripheral nodes
    
    
    return cycle, clique, star


# Write a function that will return the leading eigenvalue of the adjacency matrix of a given network.

def calculate_leading_eigenvalue(G) -> float:
    """
    Inputs:
    G: NetworkX graph object
    
    Returns:
    eig (float): leading eigenvalue of the adjacency matrix 
    
    """

    ##
    # Get the adjacency matrix as a numpy array
    A = nx.to_numpy_array(G)
    
    # Calculate the eigenvalues of the adjacency matrix
    eigenvalues = np.linalg.eigvals(A)

    
    # Return the maximum eigenvalue (leading eigenvalue)
    eig = round(max(np.real(eigenvalues)))  # Take the real part to avoid complex results
    
    
    return eig

# Example usage:
cycle, clique, star = create_toy_graphs()
cycle_eigenvalue = calculate_leading_eigenvalue(cycle)
clique_eigenvalue = calculate_leading_eigenvalue(clique)
star_eigenvalue = calculate_leading_eigenvalue(star)

# Print eigenvalues
print(f"Cycle Network Eigenvalue: {cycle_eigenvalue}")
print(f"Clique Network Eigenvalue: {clique_eigenvalue}")
print(f"Star Network Eigenvalue: {star_eigenvalue}")

    
```

What is the relationship between the eigenvalue of the adjacency matrix, the maximum degree, and average degree of each network?

邻接矩阵的最大特征值与最大度数和平均度数之间存在密切关系。一般来说，图的连通性越强（即平均度数和最大度数越大），其最大特征值也会越大。
完全网络（Clique）的最大特征值接近节点数量 n，因为它的连接性极强。
循环网络（Cycle）的最大特征值较小，因为它的每个节点只与相邻节点连接，连接性较低。
星形网络（Star）的最大特征值比循环网络大，但比完全网络小，受中心节点度数的影响较大。

```python
# Implement your code and show the visualization of Question 4
def cities_within_100(G, city_list) -> nx.Graph:
    '''
    Input
    G: NetworkX graph object
    city_list: list of strings (names of cities in G)

    Output
    S: NetworX graph object (subgraph of G that only contains edges between cities in “city_list” and directly neighboring cities that are less than 100 miles away)
    '''

    # This is a placeholder
    S = nx.Graph()

    # 遍历city_list中的每个城市
    for city in city_list:
        # 获取与当前城市相连的所有城市及其边数据
        for neighbor, data in G[city].items():
            # 如果该城市与邻居之间的距离小于100英里
            if data.get('weight', 0) < 100:
                # 将该边和两个城市添加到子图中
                S.add_edge(city, neighbor, weight=data['weight'])
               
                
    return S

```

```python
# calculate the shortest path length between each pair of nodes in the graph.
def calculate_shortest_paths(G) -> List:
    """
    Inputs
    G: NetworkX graph object

    Returns:
    shortest_paths (list): list of shortest paths between each pair of nodes in the graph.

    NOTE: 
    You may use any structure you'd like to store the shortest paths. The list return above is merely a suggestion.
    It can be a list of path lengths, a dictionary with node pairs as keys and path lengths as values, etc.
    """
    # These lines are placeholders
    # You may use any structure you'd like to store the shortest paths.
    # It can be a list of path lengths, a dictionary with node pairs as keys and path lengths as values, etc.
    shortest_paths = dict(nx.shortest_path_length(G))
    
    return shortest_paths

# plot the distribution of shortest paths using a histogram or barplot.
def plot_shortest_paths(G) -> None:
    """
    Inputs
    G: NetworkX graph object

    Returns:
    Nothing
    """
    # Create a plot of the distribution of shortest paths.
    # you may use the calculate_shortest_paths fuction to help.

    shortest_paths = calculate_shortest_paths(G)
    
    # Flatten the list of shortest path lengths
    path_lengths = [length for node_paths in shortest_paths.values() for length in node_paths.values()]
    
    # Plot histogram of path lengths
    plt.hist(path_lengths, bins=range(1, max(path_lengths) + 1), alpha=0.75, color='blue')
    plt.title('Distribution of Shortest Path Lengths')
    plt.xlabel('Path Length')
    plt.ylabel('Frequency')
    plt.show()



    plt.show()

# find the average shortest path length of the graph.
def average_shortest_path(G) -> float:
    """
    Inputs
    G: NetworkX graph object

    Returns:
    aspl (float): average shortest path length of the graph
    """
    # This line is a placeholder
    # You may use the calculate_shortest_paths fuction to help.
    shortest_paths = calculate_shortest_paths(G)
    total_length = sum(sum(length for length in node_paths.values()) for node_paths in shortest_paths.values())
    num_paths = sum(len(node_paths) for node_paths in shortest_paths.values())
    aspl = total_length / num_paths if num_paths != 0 else 0

    return aspl

# find the maximum shortest path length of the graph.
def maximum_shortest_path(G) -> int:
    """
    Inputs
    G: NetworkX graph object

    Returns:
    mspl (int): maximum shortest path length of the graph
    """
    # This line is a placeholder
    # You may use the calculate_shortest_paths fuction to help.
    shortest_paths = calculate_shortest_paths(G)
    max_path_length = max(max(node_paths.values()) for node_paths in shortest_paths.values())
    mspl = max_path_length
    
    return mspl
```

```python
import numpy as np
from typing import List

def random_walk_stationary_distribution(G, num_steps=10000) -> dict:
    """
    Perform a random walk on the graph to calculate the stationary distribution.
    
    Inputs:
    G: NetworkX graph object
    num_steps: int, number of steps for the random walk
    
    Returns:
    stationary_distribution: dict, a dictionary with nodes as keys and their stationary probabilities as values
    """
    # Initialize visit count for each node
    visit_count = {node: 0 for node in G.nodes()}
    
    # Start from a random node
    current_node = np.random.choice(list(G.nodes()))
    visit_count[current_node] += 1

    for _ in range(num_steps):
        neighbors = list(G.neighbors(current_node))
        if not neighbors:
            continue  # Skip if no neighbors
        
        # Calculate transition probabilities based on edge weights
        weights = [G[current_node][neighbor].get('weight', 1) for neighbor in neighbors]
        total_weight = sum(weights)
        probabilities = [weight / total_weight for weight in weights]
        
        # Move to the next node based on probabilities
        current_node = np.random.choice(neighbors, p=probabilities)
        visit_count[current_node] += 1

    # Normalize the visit counts to probabilities
    total_visits = sum(visit_count.values())
    stationary_distribution = {node: count / total_visits for node, count in visit_count.items()}
    
    return stationary_distribution

def popular_characters(G) -> List[str]:
    """
    Inputs:
    G: NetworkX graph object
    
    Returns:
    top_3: List[str]: a list of the string names of the 3 most commonly mentioned characters in the novel
    """
    # Calculate the stationary distribution
    stationary_distribution = random_walk_stationary_distribution(G)
    
    # Sort nodes by their stationary probability (descending order)
    sorted_nodes = sorted(stationary_distribution.items(), key=lambda x: x[1], reverse=True)
    
    # Extract the top 3 characters
    top_3 = [node for node, _ in sorted_nodes[:3]]
    
    return top_3
```

```python
# Write a function to find the weakly connected components of this network, and return how many there are and what percentage of the nodes are in the largest weakly connected component.
def weakly_connected(G) -> Tuple[int, float]:
    """
    Find the weakly connected components of the graph and return:
    - The number of weakly connected components
    - The percentage of nodes in the largest weakly connected component
    
    Inputs:
    G: NetworkX graph object
    
    Returns:
    wccs (int): number of weakly connected components in the graph
    wpct (float): percent of nodes in the graph that belong to the largest weakly connected component
    """
    try:
        # Get all weakly connected components
        components = list(nx.weakly_connected_components(G))
        
        # Number of weakly connected components
        wccs = len(components)
        
        # Size of the largest weakly connected component
        largest_wcc_size = max(len(component) for component in components)
        
        # Total number of nodes in the graph
        total_nodes = len(G.nodes())
        
        # Percentage of nodes in the largest weakly connected component
        wpct = (largest_wcc_size / total_nodes) * 100
        
        return wccs, wpct
    except Exception as e:
        print(f"Error calculating weakly connected components: {e}")
        return 0, 0.0
```

```python
# Write a function to find the strongly connected components of this network, and return how many there are and what percentage of the nodes are in the largest strongly connected component

def strongly_connected(G) -> Tuple[int, float]:
    """
    Inputs
    G: NetworkX graph object
    
    Returns:
    sccs (int): number of strongly connected components in the graph
    spct (float): percent of nodes in the graph that belong to the largest strongly connected component
    
    """
    
    #These lines are placeholders
    try:
        # Find strongly connected components using NetworkX method
        sccs_list = list(nx.strongly_connected_components(G))
        
        # Number of strongly connected components
        sccs = len(sccs_list)
        
        # Find the size of the largest strongly connected component
        largest_scc_size = max(len(component) for component in sccs_list)
        
        # Total number of nodes in the graph
        total_nodes = len(G.nodes())
        
        # Calculate the percentage of nodes in the largest SCC
        spct = (largest_scc_size / total_nodes) * 100
        
        return sccs, spct
    except Exception as e:
        print(f"Error calculating strongly connected components: {e}")
        return 0, 0.0
    
```

```python
# Calculate the average shortest path length and maximum shortest path length of the largest strongly connected component

def scc_shortest_paths(G) -> List:
    """
    Inputs
    G: NetworkX graph object
    
    Returns:
    shortest_paths (list): list of shortest paths between each pair of nodes in the graph.

    NOTE: 
    You may use any structure you'd like to store the shortest paths. The list return above is merely a suggestion.
    It can be a list of path lengths, a dictionary with node pairs as keys and path lengths as values, etc.
    
    """
    
    # These lines are placeholders
    # You may use any structure you like to store shortest paths.
    # Find all strongly connected components
    sccs = list(nx.strongly_connected_components(G))
    
    # Find the largest SCC by size
    largest_scc = max(sccs, key=len)
    
    # Subgraph of the largest SCC
    G_scc = G.subgraph(largest_scc)
    
    # Calculate shortest paths between each pair of nodes in the largest SCC
    shortest_paths_dict = dict(nx.shortest_path_length(G_scc))  # Unweighted shortest paths
    
    # Flatten the dictionary into a list of path lengths (ensure it's integers)
    path_lengths = [int(length) for node_paths in shortest_paths_dict.values() for length in node_paths.values()]
    
    return path_lengths


# plot the distribution of shortest paths for the scc using a histogram or barplot.
def plot_scc_shortest_paths(shortest_paths) -> None:
    """
    Inputs
    shortest_paths (list): list of shortest paths between each pair of nodes in the graph.

    Returns:
    Nothing
    """
    # Create a plot of the distribution of shortest paths.
    # you may use the scc_shortest_paths fuction to help.
    # Plot histogram of path lengths
    plt.hist(shortest_paths, bins=range(1, max(shortest_paths) + 1), alpha=0.75, color='blue')
    plt.title('Distribution of Shortest Path Lengths in SCC')
    plt.xlabel('Path Length')
    plt.ylabel('Frequency')
    plt.show()

# find the average shortest path length of the scc.
def average_scc_shortest_path(shortest_paths) -> float:
    """
    Inputs
    shortest_paths (list): list of shortest paths between each pair of nodes in the graph.

    Returns:
    aspl (float): average shortest path length of the graph
    """
    # This line is a placeholder
    # You may use the calculate_shortest_paths fuction to help.
    if len(shortest_paths) == 0:
        return 0.0
    
    # Calculate the average shortest path length
    aspl = sum(shortest_paths) / len(shortest_paths)
    
    return aspl

# find the maximum shortest path length of the scc.
def maximum_scc_shortest_path(shortest_paths) -> int:
    """
    Inputs
    shortest_paths (list): list of shortest paths between each pair of nodes in the graph.

    Returns:
    mspl (int): maximum shortest path length of the graph
    """
    # This line is a placeholder
    # You may use the calculate_shortest_paths fuction to help.
    # Return the maximum shortest path length
    mspl = max(shortest_paths)
    
    return mspl
```

```python
# Write a function that will determine if a graph is a DAG or not

def is_graph_dag(G) -> bool:
    """
    Input:
    G: NetworkX graph object
    
    Returns:
    is_dag (Boolean): is the graph directed or not
    """
    
    # This is a placeholder
    try:
        # Check if the graph is a directed acyclic graph (DAG)
        return nx.is_directed_acyclic_graph(G)
    except Exception as e:
        print(f"Error determining if the graph is a DAG: {e}")
        return False

```

```python
# Implement a function that will remove the first edge of each cycle

def remove_cycles(G) -> nx.Graph:
    """
    Input:
    G: NetworkX graph object
    
    Returns:
    G2: NetworkX graph object with cycles removed 
    
    """
    #This is placeholder
    # Make a copy of the graph to avoid modifying the original graph
    G2 = G.copy()
    
    # Find all simple cycles in the graph
    cycles = list(nx.simple_cycles(G2))
    
    # Iterate through each cycle and remove the first edge
    for cycle in cycles:
        if len(cycle) > 1:
            # Get the first edge in the cycle and remove it
            u = cycle[0]
            v = cycle[1]
            if G2.has_edge(u, v):
                G2.remove_edge(u, v)
    
    return G2
```


### 4.4

```python
# Write a function that returns the number of sources and the highest influencing node

def get_sources(G) -> Tuple[int, str]:
    """
    Inputs:
    G: NetworkX graph object
    
    Returns:
    num_sources (int): number of source nodes
    top_source (str): the name of the node with the highest level of influence
    """
    
    #These are placeholders
     # Find source nodes (nodes with no incoming edges)
    num_sources = sum(1 for node in G.nodes() if G.in_degree(node) == 0)
    
    # Find the node with the highest out-degree (most outgoing edges)
    top_source = max(G.nodes(), key=lambda node: G.out_degree(node))
    
    return num_sources, top_source
```

### Part 5 - Bipartite Projects of Github

```python
def load_github_data() -> Tuple[nx.Graph, List[str], List[str]]:
    """
    Returns:
    G: NetworkX graph object
    uid_list (list): list of users
    pid_list (list): list of projects
    """
    G = nx.Graph()

    # Using sets to track users and projects
    uid_set = set()
    pid_set = set()

    try:
        # Assuming the file path is correct
        file_path = "/Users/zklou/Desktop/Assignment1/github_data.txt"

        # Open and read the file line by line
        with open(file_path, "r") as file:
            for line in file:
                # Parse the user and project pair from each line
                user, project = line.strip().split()

                # Add user and project to the sets (no need to check for duplicates)
                uid_set.add(user)
                pid_set.add(project)

                # Add an edge between the user and the project
                G.add_edge(user, project)

        # Convert sets to lists
        uid_list = list(uid_set)
        pid_list = list(pid_set)

        return G, uid_list, pid_list

    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return G, [], []

    except Exception as e:
        print(f"Error loading GitHub data: {e}")
        return G, [], []
```

### 5.2

```python
def calculate_projections(G, uid_list, pid_list, device='cpu') -> Tuple[torch.sparse.Tensor, torch.sparse.Tensor]:
    """
    Inputs:
    G: NetworkX graph object
    uid_list (list): list of users
    pid_list (list): list of projects
    device (str): 'cpu' or 'cuda' to specify the device for computations
    
    Returns:
    user_matrix (torch.sparse.Tensor): one mode projection for users
    project_matrix (torch.sparse.Tensor): one mode projection for projects
    """
    
    # Initialize lists for user-user and project-project projections
    file_path = "/Users/zklou/Desktop/Assignment1/github_data.txt"
    user_indices = []
    user_values = []
    project_indices = []
    project_values = []

    # Create dictionaries to map user and project indices
    user_index = {uid_list[i]: i for i in range(len(uid_list))}
    project_index = {pid_list[i]: i for i in range(len(pid_list))}
    
    # Iterate over the edges in the graph to populate the matrices
    for user, project in G.edges():
        if user not in user_index or project not in project_index:
            continue
        
        user_idx = user_index[user]
        project_idx = project_index[project]

        # Update the user-user projection matrix (count shared projects)
        for other_user in G.neighbors(user):
            if other_user not in user_index:
                continue
            other_user_idx = user_index[other_user]
            if user_idx != other_user_idx:
                user_indices.append([user_idx, other_user_idx])
                user_values.append(1)
        
        # Update the project-project projection matrix (count shared users)
        for other_project in G.neighbors(project):
            if other_project not in project_index:
                continue
            other_project_idx = project_index[other_project]
            if project_idx != other_project_idx:
                project_indices.append([project_idx, other_project_idx])
                project_values.append(1)
    
    # Convert lists to tensors for faster operations
    user_indices = torch.tensor(user_indices, dtype=torch.long, device=device)
    user_values = torch.tensor(user_values, dtype=torch.float, device=device)
    project_indices = torch.tensor(project_indices, dtype=torch.long, device=device)
    project_values = torch.tensor(project_values, dtype=torch.float, device=device)

    # Create sparse matrices using torch.sparse
    user_matrix = torch.sparse_coo_tensor(
        indices=user_indices.t(),
        values=user_values,
        size=(len(uid_list), len(uid_list)),
        device=device
    )
    
    project_matrix = torch.sparse_coo_tensor(
        indices=project_indices.t(),
        values=project_values,
        size=(len(pid_list), len(pid_list)),
        device=device
    )
    
    return user_matrix, project_matrix
```

### 5.3

```python
# Write a function that will return the pair of users that share the highest number of Github projects between them.

def get_user_pair(M, uid_list) -> Tuple[str, str]:
    """
    Inputs:
    M: projected matrix
    uid_list (list): list of users
    
    Returns:
    u1 (str) - first user
    u2 (str) - second user
    """
    
    # Initialize variables to store the maximum shared projects and the corresponding user pair
    max_shared_projects = -1
    u1, u2 = "", ""
    
    # Iterate through the matrix to find the maximum shared projects between users
    for i in range(M.shape[0]):
        for j in range(i + 1, M.shape[1]):  # Avoid duplicate pairs by only checking the upper triangle of the matrix
            if M[i, j] > max_shared_projects:
                max_shared_projects = M[i, j]
                u1, u2 = uid_list[i], uid_list[j]
    
    return u1, u2
```

### 5.4

```python
# Write a function that will return the pair of projects that share the highest number of users between them.
def get_project_pair(M, pid_list) -> Tuple[str, str]:
    """
    Inputs:
    M: projected matrix
    pid_list (list): list of projects
    
    Returns:
    p1 (str) - first project
    p2 (str) - second project
    """
    
    # Initialize variables to store the maximum shared users and the corresponding project pair
    max_shared_users = -1
    p1, p2 = "", ""
    
    # Iterate through the matrix to find the maximum shared users between projects
    for i in tqdm(range(M.shape[0]), desc="Processing rows", unit="row"):
        for j in range(i + 1, M.shape[1]):  # Avoid duplicate pairs by only checking the upper triangle of the matrix
            if M[i, j] > max_shared_users:
                max_shared_users = M[i, j]
                p1, p2 = pid_list[i], pid_list[j]
    
    return p1, p2
```
