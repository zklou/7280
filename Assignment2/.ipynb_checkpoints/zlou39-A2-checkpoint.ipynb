### 1.1

```python
# Compute the Pearson correlation coefficieient for
# the degrees of adjacent nodes.
# use a one sample t-test to evaluate whether your
# answer is statistically significant.


def calculate_pvalue_and_pearson(G: nx.Graph) -> (float, float, str):
    """
    Inputs:
        G: NetworkX graph object

    Returns:
        p_value (float): p-value of one sample t-test of the correlation coefficient
        r (float):  the Pearson correlation coefficieient for the degrees of adjacent nodes
        network_type[*0-2*]: return one string coresponding to the network type. It has to be "assortative", "disassortative", or "neutral."
    """

    # This is a placeholder
    # The first thing is get the degrees before we process the pairs
    degrees = dict(G.degree())

    # See if there's degrees pairs, gain from u, v in nodes 
    degree_pairs = [(degrees[u], degrees[v]) for u, v in G.edges() if degrees[u] > 0 and degrees[v] > 0]

    # If there's no vaild pairs( empty, null) we return neutral.
    if not degree_pairs:
        return 1.0, 0.0, "neutral"

    # Put x,y as the values for pearson to calcuate r and p value 
    x, y = zip(*degree_pairs)

    # Use pearsonr to calc the value
    r, p_value = stats.pearsonr(x, y)

    # %5 is common used in hyposis as neutral
    if p_value > 0.05:  
        return p_value, r, "neutral"
    else:  
    # If the p-value is less than 5%, we determine the network type based on the value of r.
        if r > 0:
            network_type = "assortative"
        elif r < 0:
            network_type = "disassortative"
        else:
            network_type = "neutral"

    return p_value, r, network_type
```

### 1.2 

```python
def get_average_neighbor_degrees(G: nx.Graph) -> (List, List):
    """
    Given graph G, this function calculates the average node degrees for nodes with degree k.

    Inputs:
        G: NetworkX graph object

    Returns:
        k: a sequence representing the unique node degrees of nodes in G
        knn: a sequence representing the corresponding average neighbor degrees for each degree 
             in `k`. In other words, knn[i] is the average neighbor degree for nodes with degree k[i].
    """
    # This line is a placeholder
    # Ustify your answer and use a one sample t-test on the Pearson correlation coefficient to evaluate whether your answer is statistically significant.

    # Store the degess into default dict as a list here 
    degree_to_neighbor_degrees = defaultdict(list)

    # Then go thru all nodes fidn the avg neighbor degree
    for node in G.nodes():
        # Using the curr nodes and check if its zero
        node_degree = G.degree(node)  
        if node_degree == 0:
            continue  
            # Skip thoes nodes that are zero!

        # After that gain neighbour degree 
        neighbor_degrees = [G.degree(neighbor) for neighbor in G.neighbors(node)]
        # Calc the avg value by sum/len
        avg_neighbor_degree = sum(neighbor_degrees) / len(neighbor_degrees)  

        # Store to dict again
        degree_to_neighbor_degrees[node_degree].append(avg_neighbor_degree)

    # Calc all those k's avg neghbor degree knn by using the average neighbor degree forum
    k = sorted(degree_to_neighbor_degrees.keys())  
    knn = [sum(degree_to_neighbor_degrees[d]) / len(degree_to_neighbor_degrees[d]) for d in k]

    return k, knn
```

```python
def plot_average_neighbor_degree(k: List, knn: List):
    """
    Inputs:
        k: dictionary keys
        knn: dictionary values
    
    Returns:
        Nothing, but draws the visualizations of the graphs.
    """
    
    # This line is a placeholder
    # Set the graph with x,y labels and scales as log, also features like color and so on
    plt.figure(figsize=(6, 4))
    plt.scatter(k, knn, color="blue", alpha=0.7, edgecolors="k", label="Avg Neighbor Degree")
    plt.xscale("log")  
    plt.yscale("log")  
    plt.xlabel("Degree k")
    plt.ylabel("Average Neighbor Degree knn(k)")
    plt.title("Friendship Paradox: Degree vs. Avg Neighbor Degree")
    # I add the grid so we can better see the value of the nodes 
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)

    # Plot create 
    plt.legend()
    
    # Plot show
    plt.show()
```

```python
def average_neighbor_degree_test(k: List, knn: List) -> (float, float, str):
    """
    Inputs:
        k: dictionary keys
        knn: dictionary values
    
    Returns:
        p_value (float): p-value of one sample t-test of the correlation coefficient
        r (float):  the Pearson correlation coefficieient 
        network_type[*0-2*]: return one string coresponding to the network type

    """
    
    # This line is a placeholder
    # Before we calc the k and knn, if we only have less than 2 set of dataset, we cannot calc the network type, return netral here.
    if not k or not knn or len(k) < 2:
        return 1.0, 0.0, "neutral"

    # Calc p value and r by pearsonr
    r, p_value = stats.pearsonr(k, knn)
    # print(p_value)

    # First see the p value, if its greater than 0.05, just return neurtal as above says.
    if p_value >= 0.05:
        return p_value, r, "neutral"
    else:
        # P-value < 0.05，we need to consider the r value and check the network type
        if r > 0:
            network_type = "assortative"  # Assortative
        elif r < 0:
            network_type = "disassortative"  # Disassortative
        else:
            network_type = "neutral"  # R == 0，neutral

    return p_value, r, network_type

    
```

##### 3, p_value < 0.05 and r < 0   "disassortative"	High-degree nodes connect to low-degree nodes.

```python
# Plot the out-degree distribution
def out_degrees(G: nx.Graph) -> List:
    """
    Inputs:
        G: NetworkX graph object
    
    Returns:
        out_degrees: a list of out degrees
    """
    # This line is a placeholder
    if isinstance(G, nx.DiGraph):  # Check if it's a directed graph (had issue threre be4)
        return [G.out_degree(node) for node in G.nodes()]
    else:  # If it's undirected, use degree
        return [G.degree(node) for node in G.nodes()]

def in_degrees(G: nx.Graph) -> List:
    """
    Inputs:
        G: NetworkX graph object
    
    Returns:
        in_degrees: a list of in degrees
    """
    # This line is a placeholder
    return [G.in_degree(node) for node in G.nodes()]
```

```python
def plot_degrees(degrees: List):
    """
    Inputs:
        Degrees: list of degrees
    
    Returns:
        Nothing, but draws the visualizations of the graphs.
    """
    # This line is a placeholder
    # First we calc the degrees and probo by p(k) = count(k) / sum
    unique_degrees, counts = np.unique(degrees, return_counts=True)
    probabilities = counts / sum(counts)  
    # Here's the fig size and how the plots
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # Fig 1: Linear Scale Histogram
    axes[0, 0].bar(unique_degrees, probabilities, color="blue", alpha=0.7, edgecolor="black")
    axes[0, 0].set_xlabel("k")
    axes[0, 0].set_ylabel("p(k)")
    axes[0, 0].set_title("LINEAR SCALE")
    axes[0, 0].grid(axis="y", linestyle="--", linewidth=0.5)

    # Fig 2: Linear Binning
    bins = np.linspace(min(degrees), max(degrees), 20)
    bin_counts, bin_edges = np.histogram(degrees, bins=bins, density=True)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    axes[0, 1].scatter(bin_centers, bin_counts, color="blue", alpha=0.7)
    axes[0, 1].set_xscale("log")
    axes[0, 1].set_yscale("log")
    axes[0, 1].set_xlabel("k")
    axes[0, 1].set_ylabel("p(k)")
    axes[0, 1].set_title("LINEAR BINNING")
    axes[0, 1].grid(True, which="both", linestyle="--", linewidth=0.5)

    # Fig 3: Log Binning
    
    # First create a logspace for bin to seprate degrees into different parts
    log_bins = np.logspace(np.log10(min(degrees) + 1), np.log10(max(degrees)), num=20)
    # Calc the log-scale plot and return the freq
    log_bin_counts, log_bin_edges = np.histogram(degrees, bins=log_bins, density=True)
    # Then calc center points for all bin
    log_bin_centers = (log_bin_edges[:-1] + log_bin_edges[1:]) / 2
    axes[1, 0].scatter(log_bin_centers, log_bin_counts, color="blue", alpha=0.7)
    axes[1, 0].set_xscale("log")
    axes[1, 0].set_yscale("log")
    axes[1, 0].set_xlabel("k")
    axes[1, 0].set_ylabel("p(k)")
    axes[1, 0].set_title("LOG-BINNING")
    axes[1, 0].grid(True, which="both", linestyle="--", linewidth=0.5)

    # Fig 4: Cumulative Distribution
    # Sort degrees by npsort
    sorted_degrees = np.sort(degrees)
    cumulative_counts = np.arange(1, len(sorted_degrees) + 1) / len(sorted_degrees)
    axes[1, 1].plot(sorted_degrees, 1 - cumulative_counts, marker="o", linestyle="--", color="blue")
    axes[1, 1].set_xscale("log")
    axes[1, 1].set_yscale("log")
    axes[1, 1].set_xlabel("k")
    axes[1, 1].set_ylabel("p(k)")
    axes[1, 1].set_title("CUMULATIVE")
    axes[1, 1].grid(True, which="both", linestyle="--", linewidth=0.5)

    plt.tight_layout()
    plt.show()
```

```python
# Find the largest strongly connected component of the directed network
def largest_SCC(G: nx.Graph) -> nx.Graph:
    """
    Inputs:
        G: NetworkX graph object
    
    Returns:
        G_0: NetworkX graph object, the LCC undirected 
    
    """
    
    # This is a placeholder
    # Find all strong connects 
    sccs = list(nx.strongly_connected_components(G))
    
    # Choose the max sccs
    largest_scc = max(sccs, key=len)

    # Abs the subgraph and make a copy 
    G_scc = G.subgraph(largest_scc).copy()

    # Turn to undircected
    G_0 = G_scc.to_undirected()
    
    return G_0
```

### 3.2

```python
def create_random_network(G: nx.Graph) -> (nx.Graph, int, int):
    """
    Inputs:
        G: NetworkX graph object

    Returns:
        R:  A random NetworkX graph object
        n:  n of the G(n,p)
        p:  p of the G(n,p)

    """
    # This is a placeholder
    # Get the nodes and edges as n and m
    n = G.number_of_nodes()
    m = G.number_of_edges()

    # Calc p = 2m / (n(n-1))，create G(n, p)
    # If we have it 0 then just direct p as zero
    p = (2 * m) / (n * (n - 1)) if n > 1 else 0  

    # Create G(n,p) random map there
    # lol, the renyi is in chinese means random
    R = nx.erdos_renyi_graph(n, p)
    return R, n, p

    # Plot the clustering coefficient of the LCC and a G(n,p) network in CCDF format
def get_ccdf(values: List) -> (Tuple, List):
    """
    Inputs:
        values: clustering coefficient values

    Returns:
        unique_value: tuple
        ccdf: array or list

    """
    # This is a placeholder
    if not values:
        return np.array([]), np.array([])

    # Calc the unique vale and count
    values = np.array(values)
    unique_value, counts = np.unique(values, return_counts=True)

    # Then calc CCDF by（1 - CDF） 
    ccdf = 1 - np.cumsum(counts) / sum(counts)

    # make sure our CCDF value is in desc order
    sorted_indices = np.argsort(unique_value)[::-1]
    return unique_value[sorted_indices], ccdf[sorted_indices]


def get_clustering_coefficients(G: nx.Graph, R: nx.Graph) -> (Dict, Dict):
    """
    Compute the clustering coefficients for graphs G and R.
    The results are put into two dicts: `clustering_coefficient` and `clustering_coefficient_r`.

    Inputs:
        G: NetworkX graph object
        R: A random NetworkX graph object

    Returns:
        clustering_coefficient: dict - holds the clustering coefficients for graph G. The keys 
                                       are the nodes and the values are the corresponding clustering 
                                       coefficients.
        clustering_coefficient_r: dict - holds the clustering coefficients for random graph R.

    """
    # Using clustering to calc the coefficient for G and R
    clustering_coefficient = nx.clustering(G)
    clustering_coefficient_r = nx.clustering(R)

    return clustering_coefficient, clustering_coefficient_r


def plot_ccdf(G: nx.Graph,
              get_ccdf: Callable, # function get_ccdf
              clustering_coefficient: Dict,
              clustering_coefficient_r: Dict):
    """
    Inputs:
        G: NetworkX graph object
        get_ccdf: ccdf function previously implemented
        clustering_coefficient: dict
        clustering_coefficient_r:dict of random

    Returns:
        Nothing, but draws the visualizations of the graphs.

    """

    # This is a placeholder
    # Using list to sort the coeffienct values
    clustering_values = list(clustering_coefficient.values())
    clustering_values_r = list(clustering_coefficient_r.values())

    # Then calc CCDF nuique value
    unique_values, ccdf = get_ccdf(clustering_values)
    unique_values_r, ccdf_r = get_ccdf(clustering_values_r)

    # Drew the ccdf plot
    # Ya, those colors, haha
    plt.figure(figsize=(6, 4))
    plt.plot(unique_values, ccdf, color="#8B4513", label="Empirical Network G0")
    plt.plot(unique_values_r, ccdf_r, linestyle="--", color="#BDB76B", label="Random Network R")

    plt.xlabel("Clustering Coefficient")
    plt.ylabel("CCDF")
    plt.title("Clustering Coefficient CCDF")
    # Create and show, also grid use -- for beter view 
    plt.legend()
    plt.grid(True, linestyle="--", linewidth=0.5)
    plt.show()
```

```python
import scipy.stats as stats
import scipy as scipy

def KS_test(clustering_coefficient: Dict,
            clustering_coefficient_r: Dict) -> (scipy.stats.kstest):
    """
    Inputs:
        clustering_coefficient: dict
        clustering_coefficient_r:dict of random
    
    Returns:
        ks: KstestResults object
    
    """
    
    # This is a placeholder
    # list out coefficient values
    clustering_values = list(clustering_coefficient.values())
    clustering_values_r = list(clustering_coefficient_r.values())

    # Kolmogorov-Smirnov Test by ks_2samp, double tails
    ks = stats.ks_2samp(clustering_values, clustering_values_r)
    
    return ks

```

```python
# Plot the average clustering coefficient as a function of the node degree
def average_metric_on_degree(G: nx.Graph, metric: Dict) -> Dict:
    """
    Given a graph G and a dict `metric` that contains some metric value by node, compute the
    average metric value by node degree.
    
    Inputs:
        G: NetworkX graph object
        metric: A `metric` dict, where the keys are the nodes and the values are the metric values of the nodes. 
                An example of the metric is the cluster coefficient.
        
    Returns:
        k_coef: dict - the keys are the node degrees and the values are the corresponding average metric values
    
    """
    degree_to_metrics = {}

    # First go thru all nodes 
    for node in G.nodes():
        degree = G.degree(node)  # gain curr nodes and go thru the check 
        if degree not in degree_to_metrics:
            degree_to_metrics[degree] = []
        degree_to_metrics[degree].append(metric[node])  # store the value in metric

    # Calc the avg value of k by using the average metric value (sum/len), could work as it should 
    k_coef = {k: sum(values) / len(values) for k, values in degree_to_metrics.items()}
    return k_coef


def plot_average_clustering_coefficient(G_0: nx.Graph,
                                        average_metric_on_degree: Callable,
                                        R: nx.Graph):
    """
    Inputs:
        G: NetworkX graph object
        average_metric_on_degree: function you implemented
        R: Random NetworkX graph object
    
    Returns:
        Nothing, but draws the visualizations of the graphs. 
    
    """
    # Compute G0 and R' clustering coefficients 
    clustering_coefficient_G0 = nx.clustering(G_0)
    clustering_coefficient_R = nx.clustering(R)

    # Compute degrees' average clustering coefficient 
    k_coef_G0 = average_metric_on_degree(G_0, clustering_coefficient_G0)
    k_coef_R = average_metric_on_degree(R, clustering_coefficient_R)

    # Avoid log scale errors and using the items to store the vales of G0 and R
    k_coef_G0 = {k: v for k, v in k_coef_G0.items() if k > 0}  
    k_coef_R = {k: v for k, v in k_coef_R.items() if k > 0}

    # Extract data for plotting
    k_values_G0, coef_values_G0 = zip(*sorted(k_coef_G0.items()))
    k_values_R, coef_values_R = zip(*sorted(k_coef_R.items()))

    #Plot label and features 
    plt.figure(figsize=(8, 6))
    plt.scatter(k_values_G0, coef_values_G0, color="blue", label="Empirical Network (G0)")
    plt.scatter(k_values_R, coef_values_R, color="green", label="Random Network G(n,p)")
    
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("Degree k")
    plt.ylabel("Average Clustering Coefficient")
    plt.title("Average Clustering Coefficient vs. Degree")
    # Plot and show the graph also using -- for better value views 
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()
```

```python
def get_diameters_and_aspls(R_graphs: List) -> (List, List):
    """
    Inputs:
        R_graphs: list of random graphs
    
    Returns:
        diameters: list of corresponding diameters
        aspls: list of coresponding aspls
    
    """
    
    # This is a placeholder
    diameters = []
    aspls = []

    for R in R_graphs:
        # make sure the map is connected before calc dia and aspl
        if nx.is_connected(R):
            # Diameter calc by nx.dia
            diameters.append(nx.diameter(R))
            # Calc the aspl by nx.aspl
            aspls.append(nx.average_shortest_path_length(R))
        else:
            # If the map is not connect, just and inf to make those nodes influence result less as possible 
            diameters.append(float('inf'))
            aspls.append(float('inf'))
    
    return diameters, aspls

```

```python
def diameter_ttest(G: nx.Graph, diameters: List) -> scipy.stats.ttest_ind:
    """
    Inputs:
        G: NetworkX graph object
        diameters: list of diameters
    
    Returns:
         ttest: TtestResult object
    
    """
    
    # This is a placeholder
    # Calc lcc's dia by nxdia
    if nx.is_connected(G):
        diameter_LCC = nx.diameter(G)
    else:
        # inf the value if not connected
        diameter_LCC = float('inf')  

    # use t test for LCC, G(n, p) by dia
    ttest = stats.ttest_1samp(diameters, diameter_LCC)

    return ttest
```

```python
def aspls_ttest(G: nx.Graph, aspls: List) -> scipy.stats.ttest_ind:
    """
    Inputs:
        G: NetworkX graph object
        aspls: list of aspls
    
    Returns:
        ttest: TtestResult object
    
    """
    
    # This is a placeholder
    # Calc lcc's dia by nxdia
    if nx.is_connected(G):
        aspl_LCC = nx.average_shortest_path_length(G)
    else:
        #Inf the value if not connected
        aspl_LCC = float('inf')  

    # fliter those inf or not connceted nodes in lcc
    aspls = [x for x in aspls if x != float('inf')]

    # use t test for LCC, G(n, p) by dia
    ttest = stats.ttest_1samp(aspls, aspl_LCC)

    return ttest
```

```python
def get_LCC(G: nx.Graph) -> nx.Graph:
    """
    Inputs:
        G: NetworkX graph object
    
    Returns:
        lcc: Largest Connected Component (Directed) of G
    
    """
    
    # This is a placeholder
    # Make all tgose scc into sccs as list
    sccs = list(nx.strongly_connected_components(G))

    # Pick the max value 
    largest_scc = max(sccs, key=len)

    # Absract the subgraoh and make a copy 
    lcc = G.subgraph(largest_scc).copy()
  
    return lcc
```

```python
def get_motif_59(G: nx.Graph) -> (set, set):
    """
    Inputs:
        G: NetworkX graph object
    
    Returns:
        type_5_motif: set of type 5 motifs
        type_9_motif: set of type  motifs
    
    """

    # This is a placeholder
    # If the graph is empty, set it to ()
    if G.number_of_nodes() == 0:
        return set(), set()
        
    type_5_motif = set()
    type_9_motif = set()

    # Go thru all those maps with three nodes 
    for u in G.nodes():
        for v in G.successors(u):
            if u == v:  
                continue
            for w in G.successors(v):
                if w == u or w == v:  
                    continue

                # Ensure the uniqueness of triads not going to calc (w, v, u) 
                if not (u < v < w):
                    continue

                # Make subgraph for  (u, v, w) 
                # Create the subgraph with u,v,w
                subgraph = G.subgraph([u, v, w])

                # Count the edges in sugbgraph
                edge_count = subgraph.number_of_edges()

                # Type-5:  three edges can make a chain like (u -> v -> w)
                # So we use and, in 
                if edge_count == 2 and (u, v) in G.edges and (v, w) in G.edges:
                    type_5_motif.add((u, v, w))

                # Type-9: three edges can make a closed cricle like (u -> v -> w -> u)
                if edge_count == 3 and (u, v) in G.edges and (v, w) in G.edges and (w, u) in G.edges:
                    type_9_motif.add((u, v, w))

    
    return type_5_motif, type_9_motif
```

```python
# Number of motifs in empirical networks
def get_empirical_motifs(get_motif_59: Callable,
                         lcc: nx.Graph) -> (int, int):
    """
    Inputs:
        get_motif_59: function you already implemented
        lcc: NetworkX graph object
    
    Returns:
        type_5_motif: int - number of type 5 motifs
        type_9_motif: int - number of type 9 motifs
    
    """
    
    # This is a placeholder
    # If the graph is empty, set it to ()
    if lcc.number_of_nodes() == 0:
        return 0, 0

    
    # Calc Type-5 and Type-9 Motifs by using get_motif_59
    type_5_motif, type_9_motif = get_motif_59(lcc)

    # Get the total length for 5 and 9
    num_5_motif = len(type_5_motif)
    num_9_motif = len(type_9_motif)

    return num_5_motif, num_9_motif


# Number of motifs in random networks
def get_random_motifs(get_motif_59: Callable,
                      lcc: nx.Graph) -> (List, List):
    """
    Inputs:
        get_motif_59: function you already implemented
        lcc: NetworkX graph object
    
    Returns:
        type_5_motifs: list - the list is of size N, where N is the number of 
                       randomly generated graphs. The elements are the number 
                       of type 5 motifs for the N generated random graphs.
        type_9_motifs: list - same as above for type 9 motifs
    
    """
    
    # This is a placeholder
    N: int = 100
    num_5_motifs = []
    num_9_motifs = []

    # Get the lcc's nodes and edges
    n = lcc.number_of_nodes()
    m = lcc.number_of_edges()

    for _ in range(N):  
        # Create a random G(n, p) network with same nodes and egdes.
        # Make sure the p value in range, if n<0, just return 0 as its not in range
        p = (2 * m) / (n * (n - 1)) if n > 1 else 0  
        random_graph = nx.gnp_random_graph(n, p, directed=True)

        # Calc the Motifs
        type_5_motif, type_9_motif = get_motif_59(random_graph)

        # Doc Motifs numers 
        num_5_motifs.append(len(type_5_motif))
        num_9_motifs.append(len(type_9_motif))

    return num_5_motifs, num_9_motifs
```

```python
def motif5_ttest(num_5_motifs: List, num_5_motif: List):
    """
    Inputs:
        num_5_motifs:
        num_5_motif:
    
    Returns:
        ttest: TtestResult object
    
    """
    
    # This is a placeholder
    num_5_motifs = [x for x in num_5_motifs if x != float('inf')]

    # Check if the values are vaild for num5, return nan if the value is invaild, that not able for t test 
    if len(set(num_5_motifs)) <= 1:
        # print("Might get trouble: Type-5 Motif values are nearly identical, t-test may not be reliable.")
        return stats.TtestResult(statistic=float('nan'), pvalue=float('nan'), df=len(num_5_motifs)-1)

    # Use t test for num5
    ttest = stats.ttest_1samp(num_5_motifs, num_5_motif)

    return ttest
```

## Part 5 - Transitivity and the Average Clustering Coefficient

#### Note that the Transitivity and the Average Clustering Coefficient are two different metrics. They may often be close but there are also some extreme cases in which the two metrics give very different answers.
#### To see that, consider a network in which two nodes A and B are connected to each other as well as to every other node. There are no other links. The total number of nodes is n. What would be the transitivity and average clustering coefficient in this case (you can simplify by assuming that n is quite large)? Points will only be awarded for a mathematical derivation, however you may use code to verify your result.

##### Transitivity is defined as: T = (3 × number of triangles)/ number of connected triples
##### number of triangles = n - 2
##### number of connected triples, we will assume A, B and rest.
##### Node A can form triples with any pair i, j from the n-2 nodes: hence A = ((n-1)(n-2))/2
##### Node B can form triples with any pair i, j from the n-2 nodes: hence B = ((n-1)(n-2))/2
##### Rest nodes: (n-2)x1 = n-2
##### So the number of connected triples will be (n-1)(n-2) + (n-2)
##### The Transitivity = (3x(n-2))/(n-1)(n-2)+(n-2) = (3(n-2))/(n²-2n) = 3/n

##### Average clustering coefficient is defined as: C avg = 1/n x (from i = 1 to n) by C_i
##### Clustering Coefficient of A and B, By A = ((n-1)(n-2)) / 2, B = (n-2)as possible triangles
##### Hence clustering coefficient for A and B: (n-2)/((n-1)(n-2)/2) = 2/(n-1)
##### Clustering Coefficient for Other (n-2) Nodes, Each of the n-2 nodes only connects to A and B, forming exactly 1 triangle.
##### Their clustering coefficient is C = 1
##### Then the C avg  approx = ((2x (2/(n-1)) + 1*(n-2)) / n = (4/n+n-2)/n = n/n = 1
##### So the Average clustering coefficient is 1

#### Validate using n=1000 with code below: ( n should go to ∞)

```python
import networkx as nx

def construct_network(n):
    """
    Constructs a network where two nodes (A and B) are connected to each other and 
    to all other nodes, while all other nodes are only connected to A and B.
    """
    G = nx.Graph()
    
    # Adding nodes
    for i in range(n):
        G.add_node(i)
    
    # Connecting A (node 0) and B (node 1) to each other
    G.add_edge(0, 1)

    # Connecting A and B to all other nodes
    for i in range(2, n):
        G.add_edge(0, i)
        G.add_edge(1, i)

    return G

# Create network with 1000 nodes
n = 1000
G = construct_network(n)

# Compute transitivity and average clustering coefficient
T = nx.transitivity(G)
C_avg = nx.average_clustering(G)

print(f"Transitivity: {T:.4f}")
print(f"Average Clustering Coefficient: {C_avg:.4f}")
```
